```python
#importing libraries
import pandas as pd
import numpy as np 
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier #this imports our decision tree algo
from sklearn.metrics import accuracy_score #is used to calculate how accurate your model is by comparing predicted labels with actual labels.
from sklearn import tree #to visualize or draw a tree
import matplotlib.pyplot as plt
```

```python
balance_data = pd.read_csv("C:/Users/hp/Downloads/Decision_Tree_ Dataset.csv",sep = ',',header = 0)  #sep idicates that it a comma seperated file and header = 0 means the first row consists of column names
```

```python
print("length of dataset::",(len(balance_data)))
print("shape of dataset::",balance_data.shape)
```

```python
print("dataset::")
balance_data.head()
```

```python
balance_data.isnull().sum()
```

```python
#seperating the dependent and independent variables
X = balance_data.values[:,:4]
Y = balance_data.values[:,-1]
```

```python
#split the data 
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.3,random_state = 100)
```

```python
#training with entropy 
clf_entropy = DecisionTreeClassifier(criterion = 'entropy',random_state = 100,max_depth = 3,min_samples_leaf = 5)
clf_entropy.fit(X_train,Y_train)
```

```python
y_predict = clf_entropy.predict(X_test)
```

```python
y_predict
```

```python
print(y_predict)
print(Y_test)
```

```python
print("accuracy is",accuracy_score(y_predict,Y_test)*100)
```

```python
#HENCE THE ACCURACY OF THIS MODEL IS AROUND 93.67%
```

```python
# Visualizing the decision tree
plt.figure(figsize=(15,10))
tree.plot_tree(clf_entropy, 
               feature_names=['initial payment', 'last payment', 'credit score', 'house number'],
               class_names=['No', 'Yes'],
               filled=True, 
               rounded=True, 
               fontsize=10)
plt.savefig('decision_tree.png')  # Save the plot as an image
plt.show()
```
